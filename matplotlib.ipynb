```python
import math as mt
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=list(map(lambda x:mt.log(x),x))
plt.plot(x,y,
         label="log graph",
         color="red",
         marker="o",
         markersize=5,
         markerfacecolor="yellow",
         markeredgecolor="green",
         linewidth=2,
         linestyle="--"
         )
plt.plot(
    y,x,label="opp",
    linestyle="--",
    marker=".",
    markersize=10,
    markerfacecolor="yellow",
    markeredgecolor="green",
    linewidth=2

)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks(x,["a","b","c","d","e"],rotation=45)
plt.legend()
plt.show()
```

```python
import numpy as np
x=np.arange(0,10,0.5)
plt.figure(figsize=(5,3),dpi=200,facecolor="lightgray",num="second grapgh")
#shorthand notations
plt.plot(x,x**2,'r.--',markerfacecolor="blue",markersize=10,label="x^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.legend()
plt.show()
```

**Bar chart**

```python
labels=["a","b",'c',"d"]
value=[2,7,4,8]
color=["red","blue","black","pink"]
bars=plt.bar(
    labels,value,
    color=color,
    width=0.5,
    alpha=0.5
    )
[bar.set_hatch(h) for bar,h in zip(bars,["//","..","OO","xx"])]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.figure(figsize=(5,3),dpi=100)
plt.show()
```
