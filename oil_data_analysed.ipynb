```python
petrol=pd.read_csv("/content/Petrol Dataset.csv", encoding='latin1')
petrol.head()
```

```python
petrol.drop(["S#"],inplace=True,axis=1)
```

```python
petrol.info()
```

```python
petrol["Daily Oil Consumption (Barrels)"]=petrol["Daily Oil Consumption (Barrels)"].str.replace(",","")
petrol["Daily Oil Consumption (Barrels)"]=petrol["Daily Oil Consumption (Barrels)"].astype(float)
```

```python
petrol["World Share"]=petrol["World Share"].str.replace("%","")
petrol["World Share"]=petrol["World Share"].astype(float)
```

```python
petrol["GDP Per Capita ( USD )"]=petrol["GDP Per Capita ( USD )"].str.replace(",","")
petrol["GDP Per Capita ( USD )"]=petrol["GDP Per Capita ( USD )"].astype(float)
```

```python
petrol['Gallons GDP Per Capita Can Buy']=petrol['Gallons GDP Per Capita Can Buy'].str.replace(",","")
petrol['Gallons GDP Per Capita Can Buy']=petrol['Gallons GDP Per Capita Can Buy'].astype(float)
```

```python
petrol.info()
petrol.head()
```

```python
petrol=petrol.drop_duplicates()
```

```python
srt=petrol.sort_values(by="Gallons GDP Per Capita Can Buy",ascending=False)

srt.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(srt['Gallons GDP Per Capita Can Buy'][1:30],srt['GDP Per Capita ( USD )'][1:30],marker=".",markersize=10)
```

```python
plt.bar(petrol['Country'][:20],petrol["Daily Oil Consumption (Barrels)"][:20],label="Daily Oil Consumption (Barrels)",color="red")

#plt.bar(petrol['Country'][:20],petrol["World Share"][:20],label="World Share",color="blue")
#plt.bar(petrol["GDP Per Capita ( USD )"],label="GDP Per Capita ( USD )",color="green",height=)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Values")
plt.legend()
plt.show()
#
```

```python
plt.figure(figsize=(10, 10)) # Increased figure size
plt.pie(srt["Daily Oil Consumption (Barrels)"][10:20],
        labels=srt["Country"][10:20],
        autopct='%1.1f%%', startangle=0,
        pctdistance=0.95,explode=[.2,.2,.1,0,0,0,0,0,0,0]
        )
plt.title('Top 20 Countries by Daily Oil Consumption (Barrels)')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout() # Automatically adjusts subplot params for a tight layout
plt.show()
```
